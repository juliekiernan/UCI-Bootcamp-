# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

# PyCity School Analysis
- Charter schools outperformed District schools across all categories.  This could be due to many factors: class size, teaching style, classroom environment, and class content of courses to name a few.  Additional information would need to be provided to confirm.
- There was little discernable difference between scores of 9th graders and 12th graders.  This would seem to indicate that the success of a student is determined before they enter the 9th grade.  This could be a window of opprtunity to improve passing scores in Reading and Mathematics for the school.  If at risk/ low performing students are identified both before entry to and during the 9th grade and are provided with additional resources and educational support in these areas, their growth would ultimately contribute to the overall success for the school.
- Budget seemed to have little correlation to student success.  This could be in part to the loss of economies of gain; the larger the student population, the greater the cost of non-acedemic costs such as electricity, administration, real estate rent/ taxes, or student safety equipment.  Additional information on the breakdown of how school budgets are spent would be needed to confirm.

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total no. of schools
SchoolNames = school_data_complete_df['school_name'].unique()
NumSchools = len(SchoolNames)

#total no. of students
TotalStudents = school_data_complete_df['student_name'].count()

#total budget
TotalBudget = school_df['budget'].sum()

print(NumSchools)
print(TotalStudents)
print(TotalBudget)
```

```python
#avg math score
AvgMath = school_data_complete_df['math_score'].mean()

#avg reading score
AvgRead = school_data_complete_df['reading_score'].mean()

# % passing math, % passing reading
NumMathPass = school_data_complete_df.loc[school_data_complete_df['math_score'] >=70]['math_score'].count()
MathPassPercent = (NumMathPass/TotalStudents)*100

NumReadPass = school_data_complete_df.loc[school_data_complete_df['reading_score'] >=70]['reading_score'].count()
ReadPassPercent =(NumReadPass/TotalStudents)*100

#overall passing rate (avg of math/reading score
OverallPass = ((MathPassPercent + ReadPassPercent)/2).mean()

print(AvgMath)
print(AvgRead)
print(MathPassPercent)
print(ReadPassPercent)
print(OverallPass)
```

```python
#create dataframe
district_summary = pd.DataFrame({"Total Schools":[NumSchools], "Total Students" :[TotalStudents], 
                                 "Total Budget": [TotalBudget], 
                                 "Average Math Score": [AvgMath], "Average Reading Score": [AvgRead], 
                                 "% Passing Math": [MathPassPercent], "% Passing Reading": [ReadPassPercent], 
                                 "% Overall Passing Rate": [OverallPass]})

#clean up formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.3f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.3f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.3f}".format)
district_summary["% Passing Reading"] =district_summary["% Passing Reading"].map("{:.3f}".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map( "{:.3f}".format)

district_summary
```

```python
district_summary.to_csv("../PyCitySchools/Resources/districtsummary.csv")
```
